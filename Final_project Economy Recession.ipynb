```python
# if a column has at least one NaN , we'll select only those columns

# Loop through all columns with missing values, EXCEPT 'Target'
for col in df.columns[df.isnull().any()]:
    if col == 'Target':
        continue  #  skip the target column

    # Get non-zero, non-NaN values
    non_zero_vals = df[col][(df[col] != 0) & (~df[col].isnull())]
    
    
# assign those random values back into the DataFrame at missing positions
    if not non_zero_vals.empty:
        top_vals = non_zero_vals.value_counts().index.tolist()
        n_missing = df[col].isnull().sum()
        df.loc[df[col].isnull(), col] = np.random.choice(top_vals, size=n_missing)
    else:
        # If no valid non-zero values, fill with 0 safely
        df[col] = df[col].fillna(0)
        
df
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


# Separate the target variable from the predictors
target = df['Target']
predictors = df.drop(['DATE', 'Target'], axis=1)

# StandardScaler computes a per-feature z-score using the feature’s mean and its (population) standard deviation
# Standardize the predictors to have mean=0 and variance=1, z = x-mu / sigma
scaler = StandardScaler()
predictors_scaled = scaler.fit_transform(predictors)

# PCA check every Y with X which hass more covariance


# Perform PCA to reduce the dimensionality of the predictors
pca = PCA()
# Fit PCA to compute eigen_vectors (PCs) and eigenvalues (variances).
pca.fit(predictors_scaled)

# explained_variance_ratio_ shows fraction of total variance per PC.
explained_variance = pca.explained_variance_ratio_
# cumulative_variance is running total; 
cumulative_variance = np.cumsum(explained_variance)
# num_components finds how many PCs reach ≥95%.
num_components = np.argmax(cumulative_variance >= 0.95) + 1
print(f"Number of components needed to explain 95% of the variance: {num_components}")

# Project the data onto the top k principal components
pca = PCA(n_components=num_components)
predictors_pca = pca.fit_transform(predictors_scaled)
     
```

```python
import matplotlib.pyplot as plt
plt.plot(np.arange(1, len(explained_variance)+1), explained_variance, marker='o')
plt.xlabel('Principal Component'); plt.ylabel('Variance')
plt.title('PCA'); plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
from imblearn.over_sampling import SMOTE

def smote_and_fit(clf, X_train, y_train, X_test, y_test):
    # Oversample the training data using SMOTE
    smote = SMOTE(random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

    # Fit the classifier on the resampled (balanced) data
    clf.fit(X_train_resampled, y_train_resampled)

    # Make predictions on the test data
    y_pred = clf.predict(X_test)

    # Calculate metrics
    f1 = f1_score(y_test, y_pred, zero_division=0)
    recall = recall_score(y_test, y_pred, zero_division=0)
    precision = precision_score(y_test, y_pred, zero_division=0)

    return f1, recall, precision
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
from imblearn.over_sampling import RandomOverSampler
import matplotlib.pyplot as plt

def oversample_and_fit(clf, X_train, y_train, X_test, y_test):
    # Step 1: Oversample the training data
    oversampler = RandomOverSampler(random_state=42)
    X_train_resampled, y_train_resampled = oversampler.fit_resample(X_train, y_train)

    # Step 2: Train the classifier on oversampled data
    clf.fit(X_train_resampled, y_train_resampled)

    # Step 3: Make predictions on original test set
    y_pred = clf.predict(X_test)

    # Step 4: Calculate metrics
    f1 = f1_score(y_test, y_pred, zero_division=0)
    recall = recall_score(y_test, y_pred, zero_division=0)
    precision = precision_score(y_test, y_pred, zero_division=0)

    # Step 5: Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    tn, fp, fn, tp = cm.ravel()

    print("🔹 Confusion Matrix")
    print(f"True Positives (TP): {tp}")
    print(f"False Positives (FP): {fp}")
    print(f"True Negatives (TN): {tn}")
    print(f"False Negatives (FN): {fn}")
    print("-" * 30)

    # Step 6: Plot heatmap
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["No Recession (0)", "Recession (1)"])
    disp.plot(cmap='Blues')
    plt.title("Confusion Matrix (After Balancing)")
    plt.grid(False)
    plt.tight_layout()
    plt.show()

    return f1, recall, precision
```
